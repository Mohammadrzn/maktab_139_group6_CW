```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
p1 = (1.0, 2.0)
p2 = (3.0, 3.5)
p3 = (4.0, 6.0)
```

```python
x1, y1 = p1
x2, y2 = p2
x3, y3 = p3
```

```python
m = (y2 - y1) / (x2 - x1)
b = y1 - m * x1
```

```python
y3_pred = m * x3 + b
print(y3_pred)
```

```python
vertical_error = y3 - y3_pred
print(vertical_error)
```

```python
X = np.array([[x1], [x2], [x3]])
y = np.array([[y1], [y2], [y3]])
print(X.shape)
print(y.shape)
```

```python
theta = np.linalg.inv(X.T @ X) @ X.T @ y
print(theta)
```

```python
y_pred = X * theta
print(y_pred)
```

```python
residual = y3 - y_pred
print(residual)
```

```python
RSS = np.sum(residual ** 2)
print(RSS)
```

```python
plt.scatter(X, y)
x_line = np.linspace(1, 4, 100)
y_line = m * x_line + b
plt.plot(x_line, y_line, label="Two point line")
y_ols = x_line * theta
plt.plot(x_line, y_ols.squeeze(), color="red", label="OLS")
plt.title("Comparison of Twoâ€‘Point Line and OLS")
plt.grid(True)
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
